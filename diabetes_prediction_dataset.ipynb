```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split

# 1. Load the dataset
file_path = "/mnt/data/diabetes_prediction_dataset (1).csv"
df = pd.read_csv('/content/diabetes_prediction_dataset (1).csv')

# 2. View basic information
print("Shape of dataset:", df.shape)
print("\nColumn Info:\n", df.info())
print("\nMissing values:\n", df.isnull().sum())
print("\nFirst 5 rows:\n", df.head())

# 3. Drop duplicates if any
df = df.drop_duplicates()

# 4. Handle missing values (if any)
# You can drop, fill with median/mean, or use imputation based on your needs
df = df.dropna()

# 5. Encode categorical variables
label_encoders = {}
for col in df.select_dtypes(include=['object']).columns:
    le = LabelEncoder()
    df[col] = le.fit_transform(df[col])
    label_encoders[col] = le

# 6. Feature scaling (Standardization)
scaler = StandardScaler()
numeric_cols = df.select_dtypes(include=[np.number]).columns.tolist()

# Assume last column is target
X = df[numeric_cols[:-1]]
y = df[numeric_cols[-1]]

X_scaled = scaler.fit_transform(X)

# 7. Train-test split
X_train, X_test, y_train, y_test = train_test_split(X_scaled, y, test_size=0.2, random_state=42)

print("\nX_train shape:", X_train.shape)
print("y_train shape:", y_train.shape)
```

```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder

# 1. Load the dataset
file_path = "/mnt/data/diabetes_prediction_dataset (1).csv"
df = pd.read_csv('/content/diabetes_prediction_dataset (1).csv')

# 2. Inspect the dataset
print("âœ… Original Shape:", df.shape)
print("\nColumns:\n", df.columns)
print("\nMissing values:\n", df.isnull().sum())

# 3. Drop missing values and duplicates
df = df.dropna()
df = df.drop_duplicates()
print("âœ… Cleaned Shape (after removing missing and duplicates):", df.shape)

# 4. Fix spelling mistakes in column names (if any)
# Let's rename manually if needed; example: 'diabtes' -> 'diabetes' (adjust if needed)
df.columns = df.columns.str.strip().str.lower().str.replace(' ', '_')

# Optional: Print cleaned column names
print("\nâœ… Cleaned Column Names:\n", df.columns.tolist())

# 5. Convert object/categorical columns to meaningful numeric values
label_encoders = {}
for col in df.select_dtypes(include='object').columns:
    print(f"Encoding column: {col}")

    # Use manual mapping for known binary categorical columns
    if df[col].nunique() == 2:
        df[col] = df[col].map({'No': 0, 'Yes': 1})
    else:
        le = LabelEncoder()
        df[col] = le.fit_transform(df[col])
        label_encoders[col] = le

# 6. Final check
print("\nâœ… Data types after encoding:\n", df.dtypes)
print("\nâœ… Sample of cleaned data:\n", df.head())
```
